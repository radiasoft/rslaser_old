```python
import numpy as np
import math
```

```python
def cos_invcos_1d(phase_array):
    i_mid = int(np.size(phase_array)/2)
    z = np.cos(phase_array)
    phase_array = np.arccos(z)
    for i in range(i_mid):
        phase_array[i] *= -1.
    return phase_array
```

```python
def un_mod_the_phase_1d(phase_array):
    i_size = np.size(phase_array)
    print('i_size = ', i_size)
    diff = np.zeros(i_size)
    for i in range(i_size-1):
        diff[i] = phase_array[i+1] - phase_array[i]
    add_pi = False
    sub_pi = False
    for i in range(i_size):
        if diff[i] > 0.5*math.pi:
            add_pi = True
            sub_pi = False
        if diff[i] < -0.5*math.pi:
            add_pi = False
            sub_pi = True
        if add_pi:
            phase_array[i] += 2.*math.pi
        if sub_pi:
            phase_array[i] -= 2.*math.pi
    return diff, phase_array
```

```python
num_points = 32
phi_min = -1.4*math.pi
phi_max =  2.3*math.pi
del_phi = (phi_max - phi_min) / num_points

phi_a = np.arange(phi_min, phi_max, del_phi)
i_size = np.size(phi_a)
phi_b = np.zeros(i_size)
for i in range(i_size):
    phi_b[i] = phi_a[i_size-1-i]
phi_0 = np.concatenate((phi_a, phi_b))

phi_1 = np.mod(phi_0, math.pi)
dif_2, phi_2 = un_mod_the_phase_1d(phi_1-2*math.pi)
```
